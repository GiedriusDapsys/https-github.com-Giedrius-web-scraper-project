```python
crime.set_index('Year', inplace=True)
```

### Step 7. Delete the Total column

```python
crime.drop('Total', axis=1, inplace=True)
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
# Step 8: dešimtmečių grupavimas
# 1) Sukuriame seriją su dešimtmečiais (pvz. 1970, 1980 ir t.t.)
decades = (crime.index.year // 10) * 10

# 2) Grupuojame pagal dešimtmečius ir susumuojame skaitinius stulpelius
#    numeric_only=True užtikrina, kad nebandysime sumuoti teksto
crime_by_decade = crime.groupby(decades).sum(numeric_only=True)

# 3) Kad apsisaugotume nuo klaidų, galime išvis išmesti Population kol kas:
crime_by_decade = crime.drop('Population', axis=1) \
                     .groupby(decades) \
                     .sum()

crime_by_decade
```

### Step 9. What is the most dangerous decade to live in the US?

```python
# Step 9: Kuri dešimtmetis pavojingiausias gyventi JAV?

# 1. Sukuriame dešimtmečių seriją:
decades = (crime.index.year // 10) * 10

# 2. Išmetame „Population“ stulpelį ir susumuojame likusius pagal dešimtmečius:
crime_by_decade = (
    crime
    .drop('Population', axis=1)
    .groupby(decades)
    .sum()
)

# 3. Išrenkame dešimtmetį su didžiausiu „Violent“ (smurto) nusikaltimų skaičiumi:
most_dangerous_decade = crime_by_decade['Violent'].idxmax()

print("Pavojingiausias dešimtmetis:", most_dangerous_decade)

```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
# Step 4: Year 2061? Fix 2-digit years

# 1) Iš naujo nuskaitykite "švarią" pradinę data (jei reikia)
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
data = pd.read_csv(url, delim_whitespace=True)

# 2) Apibrėžiame f-iją pataisyti metus 61→1961:
def fix_year(y):
    return y + 1900 if y < 100 else y

# 3) Prieš drop'inant, pritaikome pataisymą:
data['Yr'] = data['Yr'].apply(fix_year)

# 4) Patikriname:
data[['Yr','Mo','Dy']].head()
```

### Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data).

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
# Nuskaitymas be antraštės
wine = pd.read_csv(url, header=None)
```

### Step 3. Assign it to a variable called wine

```python
wine.columns = [
    'Class', 'Alcohol', 'Malic_Acid', 'Ash', 'Alcalinity_of_Ash',
    'Magnesium', 'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols',
    'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280/OD315', 'Proline'
]
wine.head()
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
cols_to_drop = wine.columns[[0, 3, 6, 8, 10, 12, 13]]
wine.drop(cols_to_drop, axis=1, inplace=True)
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
import numpy as np
wine.loc[0:2, 'alcohol'] = np.nan
wine.head(5)
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.loc[[3, 4], 'magnesium'] = np.nan
wine.loc[2:6, ['magnesium']]  # show rows 2–6 in magnesium
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine['alcohol'].fillna(10, inplace=True)
wine['magnesium'].fillna(100, inplace=True)
wine.head(6)
```

### Step 9. Count the number of missing values

```python
wine.isna().sum()
```

### Step 12.  How many missing values do we have?

```python
print("Missing per column:")
print(wine.isna().sum())
print("Total missing values:", wine.isna().sum().sum())
```

```python
wine.dropna(axis=0, inplace=True)
wine.shape  # see how many rows remain
```
